```python
def toRGB(img):
    '''
    Invert R and B channel, for dislaying images as matplotlib assumes RGB and cv reads BGR
    '''
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
```

```python
def load_picture(pict_name):
    from PIL import Image
    path=os.path.sep.join([INPUT_FOLDER,pict_name])
    im = Image.open(path)
    return im
```

```python
def cropping_faces(image):
    aligned=[]
    x_aligned, prob = mtcnn(image, return_prob=True)
    for face,p in  zip(x_aligned,prob):
        #print(f'Face detected with probability: {p:.6f}')
        if p>=MIN_FACE_CONFIDENCE:
            aligned.append(face)
    return aligned
```

```python
def padding(A,final_size):
    h,w,c = A.shape
    before_x = (final_size[0]-h)//2
    after_x = final_size[0]-before_x-h
    before_y = (final_size[1]-w)//2
    after_y = final_size[1]-before_y-w
    return np.pad(A,((before_x,after_x),(before_y,after_y),(0,0)), mode='constant')
```

```python
def zoom_image(image, final_size): 
    (h,w) = image.shape[:2]
    if w > final_size[1]:
        if int(h*final_size[1]/w)>final_size[0]:
            return cv2.resize(image, (int(w*final_size[0]/h),final_size[0]))
        else:
            return cv2.resize(image, (final_size[1], int(h*final_size[1]/w)))
    elif h>final_size[0]:
        return cv2.resize(image, (int(w*final_size[0]/h),final_size[0]))
    else:
        return image
```

```python
image_list=os.listdir(IMAGE_FOLDER)
```

```python
# resizing all pictures with keeping h/w ratio
batch=61
k=0
N=int(np.ceil(len(image_list)/batch))
final_size = (800,1000)
```

```python
os.mkdir(RESIZED_IMAGE_FOLDER)
for k in range(N):
    Folder=os.path.sep.join([RESIZED_IMAGE_FOLDER, f'Folder_{k}'])
    os.mkdir(Folder)
    X=[]
    for img in image_list[k*batch:(k+1)*batch]:
        path=os.path.sep.join([IMAGE_FOLDER,img])
        image=cv2.imread(path)
        image=zoom_image(image,final_size)
        X.append(image)
    X=np.array(X)
    X_padded=[padding(img,final_size) for img in X]
    for i,image in enumerate(X_padded):
        output_path = os.path.sep.join([Folder, f'friends_{k}_{i}.png'])
        cv2.imwrite(output_path,image)
```
